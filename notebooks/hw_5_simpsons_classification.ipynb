````python
class SimpsonsDataset(Dataset):
    def __init__(self, files, label_encoder, mode):
        super().__init__()
        #    
        self.files = sorted(files)
        #  
        self.mode = mode
        if self.mode not in DATA_MODES:
            print(f"{self.mode} is not correct; correct modes: {DATA_MODES}")
            raise NameError

        self.label_encoder = label_encoder
        self.len_ = len(self.files)

    def __len__(self):
        return self.len_ #  self.__len__()   ,   .
        #         -
        #     self.__len__()

    def __getitem__(self, index):
        x = self.load_image(self.files[index])
        x = self.transform_images_to_tensors(x)

        if self.mode == 'test':
            return x
        else:
            path = self.files[index]
            y = self.label_encoder.transform([path.parent.name,]).item()
            return x, y

    #         
    def load_image(self, file):
        image = Image.open(file)
        image.load()
        return image

    #    
    def transform_images_to_tensors(self, image):
      if self.mode == 'train':
        transform = v2.Compose([
            v2.PILToTensor(),
            v2.ToDtype(torch.float32, scale=True),
            v2.Normalize(NORMALIZE_MEAN, NORMALIZE_STD),
            #v2.ColorJitter(brightness = 0.1, contrast = 0.1, saturation = 0.1),
            v2.Resize(RESCALE_SIZE, antialias=True),
          ])
      else:
        transform = v2.Compose([
            v2.PILToTensor(),
            v2.ToDtype(torch.float32, scale=True),
            v2.Normalize(NORMALIZE_MEAN, NORMALIZE_STD),
            v2.Resize(RESCALE_SIZE, antialias=True),
          ])

      tensor_transformed = transform(image)
      return(tensor_transformed)

    @property
    def classes(self):
        if hasattr(self.label_encoder, "classes_"):
            return list(self.label_encoder.classes_)
````

````python
def imshow(inp, title=None, plt_ax=plt, default=False):
    """Imshow  """
    inp = inp.numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    inp = std * inp + mean
    inp = np.clip(inp, 0, 1)
    plt_ax.imshow(inp)
    if title is not None:
        plt_ax.set_title(title)
    plt_ax.grid(False)
````

````python
class SimpleCnn(nn.Module):
    """
           .

          ,   
        ,   ReLU  
     (max-pooling).     
         .

    :
    ----------
    n_classes : int
           .

    :
    ----------
         (3, H, W), 
    3 -  rgb   ,  H  W -    ,
    .      (n_classes).

    :
    -------
    forward(x):
                
         .
    """

    def __init__(self, n_classes):
        super().__init__()
        self.conv1 = nn.Sequential(
            nn.Conv2d(in_channels=3, out_channels=8, kernel_size=3),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2)
        )
        self.conv2 = nn.Sequential(
            nn.Conv2d(in_channels=8, out_channels=16, kernel_size=3),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2)
        )
        self.conv3 = nn.Sequential(
            nn.Conv2d(in_channels=16, out_channels=32, kernel_size=3),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2)
        )
        self.conv4 = nn.Sequential(
            nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2)
        )
        self.conv5 = nn.Sequential(
            nn.Conv2d(in_channels=64, out_channels=96, kernel_size=3),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2)
        )

        self.out = nn.Linear(96 * 5 * 5, n_classes)


    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.conv3(x)
        x = self.conv4(x)
        x = self.conv5(x)

        x = x.view(x.size(0), -1)
        logits = self.out(x)
        return logits
````

````python
def fit_one_epoch(model, optim, criterion, dataloader, epoch, device):
    total = 0
    correct = 0
    running_loss = 0.0
    
    model.train()
    for data, label in dataloader:
        data, label = data.to(device), label.to(device)
        
        output = model(data)
        loss = criterion(output, label)
        loss.backward()
        optim.step()
        optim.zero_grad()
        
        bs = data.size(0)
        running_loss += loss.item() * bs
        total += bs
        pred = output.argmax(1)
        correct += pred.eq(label).sum().item()

    avg_loss = running_loss / total
    acc = 100. * correct / total
    
    return float(avg_loss), float(acc)
````

````python
def eval_one_epoch(model, criterion, dataloader, epoch, device, best_acc, model_name='model'):
    total = 0
    correct = 0
    running_loss = 0.0
    
    model.eval()
    with torch.no_grad():
        for data, label in dataloader:
            data, label = data.to(device), label.to(device)
            
            output = model(data)
            loss = criterion(output, label)
            
            bs = data.size(0)
            running_loss += loss.item() * bs
            total += bs
            pred = output.argmax(1)
            correct += pred.eq(label).sum().item()
    
    avg_loss = running_loss / total
    acc = 100. * correct / total
    
    if acc > best_acc:
        # print('Saving..')
        weights = Path(working_dir) / f'ckpt_{model_name}.pth'
        state = {
            'net': model.state_dict(),
            'acc': acc,
            'epoch': epoch,
        }
        torch.save(state, weights)
    
    return float(avg_loss), float(acc)
````

````python
import time

def train_func(model, optimizer, criterion, dataloaders, num_epochs, device, model_name):
    start_time = time.time()
    
    train_losses = []
    val_losses = []
    best_acc = 0
    best_epoch = -1
    
    for i in range(num_epochs):
        train_loss, train_acc = fit_one_epoch(model, optimizer, criterion, loaders['train'], i, device)
        train_losses.append(train_loss)
    
        val_loss, val_acc = eval_one_epoch(model, criterion, loaders['val'], i, device, best_acc, model_name)
        val_losses.append(val_loss)
        
        best_acc = max(best_acc, val_acc)
        best_epoch = i if best_acc == val_acc else best_epoch
    
        print(f"Epoch {i+1}/{num_epochs} | Train loss: {train_loss:.4f}, acc: {train_acc:.2f}% | "
              f"Validation loss: {val_loss:.4f}, acc: {val_acc:.2f}%")
    
    end_time = time.time()
    total_time = end_time - start_time
    
    print(f"\n    {total_time/60:.2f} ")
    print(f"  : {best_acc:.2f}%   {best_epoch+1}")
    
    train_losses, val_losses
    return train_losses, val_losses, best_acc, best_epoch, total_time
````

````python
import os

def train_or_load_weights(model, optimizer, criterion, dataloaders, num_epochs, device, model_name):
    weights = Path(working_dir) / f'ckpt_{model_name}.pth'

    if weights.exists():
        ckpt = torch.load(weights, map_location=device)
        model.load_state_dict(ckpt['net'], strict=True)

        print(f"    {weights} | : {ckpt['acc']:.2f}%")
        return

    train_losses, val_losses, best_acc, best_epoch, total_time = train_func(
        model, optimizer, criterion, dataloaders, num_epochs, device, model_name
    )
        
    plt.figure(figsize=(18, 9))
    plt.plot(np.arange(len(train_losses)), train_losses, label='Train loss')
    plt.plot(np.arange(len(val_losses)), val_losses, label='Validation loss')
    plt.title(f'Training summary (10 epochs, total time: {total_time:.1f}s)')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend(loc='best')
    plt.show()
````

````python
class SimpleCnnClassifier(nn.Module):
    def __init__(self, n_classes):
        super().__init__()
        self.conv1 = nn.Sequential(
            nn.Conv2d(in_channels=3, out_channels=8, kernel_size=3),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2)
        )
        self.conv2 = nn.Sequential(
            nn.Conv2d(in_channels=8, out_channels=16, kernel_size=3),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2)
        )
        self.conv3 = nn.Sequential(
            nn.Conv2d(in_channels=16, out_channels=32, kernel_size=3),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2)
        )
        self.conv4 = nn.Sequential(
            nn.Conv2d(in_channels=32, out_channels=64, kernel_size=3),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2)
        )
        self.conv5 = nn.Sequential(
            nn.Conv2d(in_channels=64, out_channels=96, kernel_size=3),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2)
        )

        self.fc1 = nn.Sequential(
            nn.Linear(96 * 5 * 5, 256),
            nn.ReLU()
        )
        self.fc2 = nn.Sequential(
            nn.Linear(256, 128),
            nn.ReLU()
        )
        
        self.out = nn.Linear(128, n_classes)


    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.conv3(x)
        x = self.conv4(x)
        x = self.conv5(x)

        x = x.view(x.size(0), -1)
        x = self.fc1(x)
        x = self.fc2(x)
        logits = self.out(x)
        return logits
````
