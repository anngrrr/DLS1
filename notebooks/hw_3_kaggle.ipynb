```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
def search_and_draw(X, y, model, param_name, grid, param_scale='ordinary', draw=True):
    """
            .

    :
    ----------
    X : array-like of shape (n_samples, n_features)
         .  .
    y : array-like of shape (n_samples,)
            .
    model : estimator object
           ,       .
    param_name : str
          ,        .
    grid : array-like
           ,   .
    param_scale : {'ordinary', 'log'}, optional (default='ordinary')
            x  .  'log',    .
    draw : bool, optional (default=True)
        , ,        .

    :
    -----------
    means : ndarray of shape (len(grid),)
                  .
    error : ndarray of shape (len(grid),)
                  .

    :
    ------------
      GridSearchCV   -    roc_auc.
      draw   True,       
        .      .
    """
    parameters = {param_name: grid}

    CV_model = GridSearchCV(
        estimator=model,
        param_grid=parameters,
        cv=5,
        scoring='roc_auc',
        n_jobs=-1,
        verbose=10
    )
    CV_model.fit(X, y)
    means = CV_model.cv_results_['mean_test_score']
    error = CV_model.cv_results_['std_test_score']

    if draw:
        tic = np.arange(len(grid))
        
        plt.figure(figsize=(10,4))
        plt.title('choose ' + param_name)


        if (param_scale == 'log'):
            plt.xscale('log')

        plt.plot(tic, means, label='mean values of score', color='red', lw=3)
        plt.fill_between(tic, means - 2 * error, means + 2 * error,
                         color='green', label='filled area between errors', alpha=0.5)
        
        legend_box = plt.legend(framealpha=1).get_frame()
        legend_box.set_facecolor("white")
        legend_box.set_edgecolor("black")

        plt.xticks(tic, [str(x) for x in grid])
        plt.xlabel('parameter')
        plt.ylabel('roc_auc')
        
        plt.show()

    return means, error
```

```python
log_reg = LogisticRegression(max_iter=1000, random_state=42)
param_name = 'C'
grid = [0.001, 0.01, 0.1, 1, 10, 100]

result = search_and_draw(
    data[num_cols_proc + cat_cols_proc],
    data[target_col],
    log_reg,
    param_name,
    grid,
);
```

```python
print(f'  `roc_auc` {result[0][2]}  ``   {grid[2]}')
```

```python
from catboost import CatBoostClassifier
from sklearn.metrics import roc_auc_score

X_train, X_valid, y_train, y_valid = train_test_split(
    data[num_cols + cat_cols],
    data[target_col],
    test_size=0.2,
    stratify=data[target_col],
    random_state=42
)

cat_bst = CatBoostClassifier(
    iterations=3,
    depth=2,
    learning_rate=1,
    loss_function='Logloss',
    verbose=False
)

cat_bst.fit(X_train, y_train, cat_features=cat_cols)

preds_proba = cat_bst.predict_proba(X_valid)[:, 1]
print("roc_auc = ", roc_auc_score(y_valid, preds_proba))
```

```python
def cat_boost_grid_search(X, y, cat_cols, param_name, grid, random_seed=42):
    X_train, X_valid, y_train, y_valid = train_test_split(
        X,
        y,
        test_size=0.2,
        stratify=y,
        random_state=random_seed
    )

    base_params = {
        'iterations': 500,
        'depth': 3,
        'learning_rate': 0.05,
        'loss_function': 'Logloss',
        'verbose': False,
        'random_seed': random_seed
    }
    
    scores = []
    for v in grid:
        params = {**base_params, param_name: v}
        cat_bst = CatBoostClassifier(**params)
        cat_bst.fit(
            X_train,
            y_train,
            cat_features=cat_cols
        )
        preds_proba = cat_bst.predict_proba(X_valid)[:, 1]
        scores.append(roc_auc_score(y_valid, preds_proba))
    return scores

def cat_boost_search_and_draw(X, y, cat_cols, param_name, grid):
    scores = cat_boost_grid_search(X, y, cat_cols, param_name, grid)
    tic = np.arange(len(grid))
        
    plt.figure(figsize=(10,4))
    plt.title('choose ' + param_name)
    plt.plot(tic, scores, label='values of score', color='red', lw=3)
    
    legend_box = plt.legend(framealpha=1).get_frame()
    legend_box.set_facecolor("white")
    legend_box.set_edgecolor("black")

    plt.xticks(tic, [str(x) for x in grid])
    plt.xlabel(param_name)
    plt.ylabel('roc_auc')
    plt.show()

    return scores
```

```python
param_name = 'depth'
grid = [1, 2, 3, 4, 5, 6]

scores = cat_boost_search_and_draw(
    data[num_cols + cat_cols],
    data[target_col],
    cat_cols,
    param_name,
    grid,
)
```

```python
param_name = 'learning_rate'
grid = [0.01, 0.02, 0.03, 0.05, 0.07, 0.1, 0.15, 0.2]

scores = cat_boost_search_and_draw(
    data[num_cols + cat_cols],
    data[target_col],
    cat_cols,
    param_name,
    grid,
)
```
